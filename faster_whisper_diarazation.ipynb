```python
import time
import torch
from transformers import pipeline
from datetime import timedelta
from pyannote.audio import Pipeline
from transformers.utils import is_flash_attn_2_available
```

```python
pipe = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-large-v3", # select checkpoint from https://huggingface.co/openai/whisper-large-v3#model-details
    torch_dtype=torch.float16,
    device="cuda:0", # or mps for Mac devices
    model_kwargs={"use_flash_attention_2": is_flash_attn_2_available()},
)
```

```python
diarazation_pipeline = Pipeline.from_pretrained("pyannote/speaker-diarization-3.0", use_auth_token="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")
```

```python
diarazation_pipeline.to(torch.device("cuda"))
```

```python
def process_audio_diarization_with_transcription(audio_file_path_mp3):
    start_time = time.time()
    diarization = diarazation_pipeline(audio_file_path_mp3)
    end_time = time.time()

    print(f"Execution time For Diarization: {end_time - start_time} seconds")

    start_time = time.time()
    transcription = pipe(audio_file_path_mp3, chunk_length_s=40, batch_size=20, return_timestamps=True)
    end_time = time.time()

    print(f"Execution time For Whisper: {end_time - start_time} seconds")


    transcript = []
    processed_chunks = set()  # To track processed chunks

    for turn, _, speaker in diarization.itertracks(yield_label=True):
        speaker_transcript = ""
        turn_start, turn_end = turn.start, turn.end

        for chunk_index, chunk in enumerate(transcription['chunks']):
            chunk_start, chunk_end = chunk['timestamp']
            # Check for overlap and if the chunk hasn't been processed already
            if chunk_start < turn_end and chunk_end > turn_start and chunk_index not in processed_chunks:
                speaker_transcript += chunk['text'] + " "
                processed_chunks.add(chunk_index)  # Mark as processed

        if speaker_transcript:
            transcript_entry = {
                "timestamp": str(timedelta(seconds=int(turn_start))),
                "speaker": f"Speaker {speaker}",
                "text": speaker_transcript.strip()
            }
            transcript.append(transcript_entry)

    return transcript, transcription, diarization
```

```python
start_time = time.time()
transcript, transcription, diarization = process_audio_diarization_with_transcription('b.mp3')
end_time = time.time()

print(f"Execution time: {end_time - start_time} seconds")
```
